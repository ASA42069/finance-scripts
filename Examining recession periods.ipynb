```python
#Recession periods
link = 'https://en.wikipedia.org/wiki/List_of_recessions_in_the_United_States#:~:text=The%20average%20recession%20lasted%2022,an%20average%20of%2057%20months.'
data = pd.read_html(link)[0]
data = data.rename(columns={"Dates[nb 2]":"Dates"})
data

# there are 2 tables in wikipedia [0] gets the intro table on top and [1] gets the bigger table below
```

```python
data2 = pd.read_html(link)[1]
del data2["Business activity [nb 3]"];del data2["Trade & industrial activity[nb 3]"]
data2 =data2.rename(columns={"Dates[nb 4]":"Dates"})
data2
```

```python
data3 = pd.read_html(link)[2]
del data3["Peak unemploy­ment"]
del data3["GDP decline (peak to trough)"]
data3 =data3.rename(columns={"Period Range":"Dates","Duration (months)": "Duration","Time since previous recession (months)": "Time since previous recession"})
data3
```

```python
frames = [data,data2,data3]
new= pd.concat(frames)
new = new.reset_index(drop=True)
new
```

```python
column_name = "recessions"
print('Do you want to save the data to excel?')
savedata_excel = str(input("Enter y or n: "))
if savedata_excel == 'y':
    file_name = str(input("Enter file name - no spaces: ")) #name the file as fit
    filename= file_name+'.xlsx'
    new.to_excel(filename,sheet_name=column_name)
    print('Data saved to excel! Download the sheet if you wish.')
elif savedata_excel == 'n':
    print()
    print('You have opted not to save the data to excel')
```
